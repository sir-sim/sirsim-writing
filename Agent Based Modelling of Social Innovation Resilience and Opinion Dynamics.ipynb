# Agent Based modelling of Social Innovation, Resilience, and Opinion Dynamics

## Abstract
The resilience literature deals primarily with natural systems. For such systems the formal language of basins of attraction within dynamical systems has yielded important insights. Differential equation models of resilience have yielded important insights about systems ranging from forest Savannah ecosystems to freshwater lakes. Frances Westley argues that social innovation can be understood similarly in terms of resilience and opportunity. 

Extending the approach to social systems has proven challenging, however. In social systems, agents struggle to retain features of a system they want and change features they don’t. They navigate towards imagined futures based on ideas they hold within their minds. To model these agents, we must model agents with minds attempting to change the system. Individuals and groups, which are clusters of individuals who are in communication and have similar opinions, decide to act based on their ideas about what is possible and what is desirable. 

The standard lumped differential equation models of resilience does not represent agents. Furthermore, existing models of agents typically don’t include agents with minds that represent opportunities and the possibility of changing systems. As a result there has not been a satisfying approach to formally modelling social innovation. 

We introduce a new agent-based approach to modelling social innovation incorporating agents that imagine change and techniques drawn form recent models of how neurons communicate in brains.



## Introduction

The resilience literature deals primarily with natural systems. For such systems the formal language of basins of attraction within dynamical systems has yielded important insights. The challenge in applying the concept of resillience to social systems is that resilience in social systems is fundamentally different from resilience in natural systems because social systems are fundamentally different from natural systems.

A key element of a social system is the agent, a decision-making element.  Agency is inherently teleological. Agency is absent from models of natural systems but central to human systems. Decision making implies intention, and ascribing intention to an agent implies necessarily that the agent has some model of the future. Agent-based models that  incorporate some representation of an agent's model of the world linked to the agent's desires therefore provide the natural approach to the study of complex social systems. 

The standard lumped differential equation models of resilience that work for natural systems can’t represent agents.

The models held by agents can be formalized as a set of state variables that modulate agent actions in response to external events. The value of the state variables at any time can be referred to as the agent's "opinion" at that point in time. Opinions may change over time. 

Agents with opinions are connected in a network of messages. Some aggregation of the messages can be seen as `public opinion.`  Aggregates over subsets of agents can be seen as group opinion.  A variety of social choice rules, ranging from pure majority rule through oligarchy to dictatorship may apply. Some subset of agents have the power to do or decide. In this framework, social change can be seen as shifts in the opinion of a decisive subset of agents.

Frances Westley argues that social innovation can be understood in terms of resilience and opportunity. "Innovators" disseminate messages containing information about opportunities: about the potential benefit of a decision and the probability of the benefit being achieved. It seems likely that "innovators" are also "enthusiasts," sending frequent signals of high intensity and actively adding links to others.

## Modelling the society

We can think of innovation as a  process beginning with change at the individual level in a subset of agents, which we can call "discovery," followed by a collective prbenefits form adoptionocess of consideration that may lead to adoption or rejection. Proponents of an innovation necessarily expect adoption to have a positive effect for themselves or more generally. 

There may,  however, be a subset of agents for whom an innovation has a negative expected value. The interests of petroleum producers, for example, are threatened by the introduction  of renewable energy and limitations on carbon emissions. These intersts may block adoption of ainnovatios that have a positve net value. 

An innovation may be rejected because it is utimately assesse to provide no net benefit relative to alternatives. A system that rejects harmful innovations can be seen as resilient.  A system that rejects beneficial innovations can also be seen as resilient, but this resilience has negative implications. 

Some innovations must be adopted at the collective level while others can be introduced locally. A model of innovation necessarily will incorporate a criterion for identifying whether an innovation is adopted or not as well as whether the innovation sis adopted locally or generally.

To model the process of adoption or rejection of an innovation we adopt from the economics and political literature the notion of a `social decision rule.` These rules include a variety of voting mechanisms, including majority rule, as well as Utilitarian and Bergson-Samuelson welfare functions that sum benefits to individuals. The rules may be applied to the entire community or locally, allowing adoption by connected subsets of indiviudals. 

Information abouut potential innovatios spreads though the comunication llinks among individuals an groups. It is well known that network structure can determine the speed of information flow, the power of groups, the influence of individuals and even economic returns to groups and individuals. Investigating the inovation process forces us to examine the role of network structure  in determinieng whether and how an innovation is adopted or rejected. 

A feature of social networks that is shared with neural networks, is that agents may dynamically adjust the strength or number of links. A neuron, like a person, can cut off commnunication from a source, discount that information, or even add new recipients. Research on neurons provides usable models for the evolution of a network and for the evolution of the ways agents respond. These adaptations represent learning at the level of the neuron and at the collective level. 

A great deal of resarch is focussed on learning in neural systems. This learning can be described as adaptation to new tasks. A similar process must happen in social systems. Our focus, however, is on the selection of new tasks -  a one-off decision process.


## Social dynamics: resilience and opportunity

If the models held by agents are essential, the dynamics of the system involve iteratively updating the opinions using signals from other agents. Updating continues until until some pre-defined institutional threshold is reached or until it becomes clear that that trigger will not be reached and the status quo will be maintained.
 
In a system that reaches equilibrium, we can imagine a stable state reached at, say, time $t$ such that 

$x_t=f(x_t)$ 

or 

$e_{it}=f(e_it) \hspace{1cm}  \forall i$ 

This is simply a description of a fixed point in the space of messages. It is a self-confirming equilibrium in the sense of  Fudenberg and Levine (Fudenberg, Drew and David K. Levine. Self-confirming Equilibrium, Econometrica, Vol. 61, No. 3 (May, 1993), 523-545)  in which  each player's beliefs are correct along the equilibrium path of play. If a self-confirming equilibrium occurs repeatedly, no player ever
observes play that contradicts her belief. 

An innovation may be adopted as a result of new information entering the network as exogenous signals to some or all agents. 

$x_1=f(x_0+u)$

The new signal can be news about a new technology or innovation. The new signal can be a broadcast signal - received by all agents at once, but more commonly new information comes from a subset of agents. We might see these agents as "broadcasters", "enthusiasts" or "innovators."

* An **idea/innovation** is of the form "do this now."  In social terms an idea is introduced as a subset of the population for whom the opinion of benefit and probability leads to strong and persistent signals.


### Experiments
Beginning with a network of agents described by a  set of directed links, a stable initial distribution of opinions, and a specific social decision rule, our central question is  "Under what conditions will an innovation be adopted?" 

The question is similar to that posed in evolutionary game theory: under what condition is a population proof against invasion by a new behavioural phenotype. In evolutionary game theory the  Evolutionarily Stable Strategy (ESS)is a population mix which is an attractor - and which is robust against the introduction of alternative strategies under a specific, evolution-like process under which strategies increase or decrease in population share.  

An experiment consists of introducing a mutant opinion in a fraction of the population and either

  1) observing some aggregate of the opinions across the network as a result of the innovation. This could be thought of as approximating a voting process. We ask if the innovation can achieve a large enough change in the pattern of opinion to achieve a positive majority for the appropriate subset of agents.

  2) observing when or if a designated agent, or "decider" reaches a critical value in opinion based on signals coming to it from it's links.
  
Rejection of an innovation is evidence of a resilient system, since innovation need not be judged positive. A resilient system, however, may in fact be one which resists positive changes. 

Several parameters of the network will affect the likelihood of rejection. these include 

  1) the connectedness of agents, 
  
  2) the degree of localization of the network,
  
  3) the existence of influential agents
  
as will the rules by which agents adjust their opinions. 




## Modelling the agent

We deal with an agent, $j$ that receives messages about a potential innovation from a set $I$ of agents and sends a message to a set $H$. Messages are 2-vectors with information about 

  1) benefit
  
  2) probability

Extensions can introduce a higher dimension vector of values. Individual $j$'s "opinion" in our model is also a 2-vector, $\mathbf{𝑒}_j$, representing $j$'s own estimates of the  benefit and the probability of implementing the innovation. This "opinion", $\mathbf{𝑒_j}$, represents a model of a possible future. `(measure of the desirability of an uncertain future.)`

The elements of $\mathbf{𝑒}_j$ are `state variables.` Both seem to be essential, since neither a high benefit from an impossible outcome nor a very low or negative benefit from a highly likely outcome is likely to gain wide support. Including both components models a system with uncertainty and probabilistic decision making. In economic terms the product of benefit and probability is known as expected value. 

$\mathbf{𝑒}_j$ is analogous to the neuron's "encoding vector" in the work of `REF XXXXXXX.` The agent will pass on most strongly information that aligned with it's opinion.

The  messages received by $j$ from the set $I$,  $\mathbf{x}_I$, are  weighted and summed by $j$ to produce a 2-vector summary that is that is commensurate with $j$'s opinion vector. 


$\hat{\mathbf{x}}_j= \mathbf{x}_I\cdot\omega$

It is then filtered though $j$'s own opinion. `NOTE: An aggregation over the population rather than just those received by j can be interpreted as collective opinion or public opinion. The agent's communication behaviour $\mathbf{a}_j$ is computed from $j$'s opinions and received messages. Since some agents respond more strongly than others we multiply the filtered input by an  individual by a  `gain` parameter, $\alpha_j$ which determines how strongly agent $j$ responds to stimulus. Since agents may be voluble or be reticent about signalling, we add a bias ,  $b_i \in (-1-1)$. The result is 

$$
\mathbf{a}_j = G(\alpha_j \mathbf{e}_j \cdot \hat{\mathbf{x}}_j + b_i) \\
$$

$G$ is a possibly  nonlinear function of a linear function of messages received.  Stewart (2012) describes a neural model homologous to ours in which $j$ receives a 2-vector signal from each member of a set $I$ of other agents. The $Ix2$ signal is weighted to extract information about the two components  and these components are filtered though the neuron's "encoding vector"  which we term "opinion"  

The signal $\mathbf{a}_j$ will generally have multiple recipients. It will vary over time in response to the messages received and $j$'s evolving opinion vector.
